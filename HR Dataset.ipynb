# DATA ANALYSIS

1. Count of employees in each department

```python
Count_Employee=data.groupby("Department")["Department"].count()
```

```python
Count_Employee
```

2. Average salary of employees in each department

```python
Avg_Salary=data.groupby("Department")["Salary"].mean()
```

```python
Avg_Salary.round(2)
```

3. Calculate the average of salary of employees based on gender.

```python
Genderbased_Salary=data.groupby(["Department","Sex"],as_index=False).Salary.mean()
```

```python
Genderbased_Salary
```

4. Most popular Recruitment Source of the company based on department 

```python
Recruitment_Source=data.groupby(["Department","RecruitmentSource"]).size().reset_index(name="counts")
```

```python
Recruitment_Source
```

5.Total number of working days for terminated employees

```python
Employee_Name=input("Enter the Employee name:")
```

```python
Start_year=int(input("Enter the start year: "))
Start_month=int(input("Enter the start month: "))
Start_day=int(input("Enter the start day: "))
Start_Date=date(Start_year,Start_month,Start_day)
Start_Date
```

```python
T_year=int(input("Enter the Terminated year: "))
T_month=int(input("Enter the Terminated month: "))
T_day=int(input("Enter the Terminated day: "))
T_Date=date(T_year,T_month,T_day)
T_Date
```

```python
business_days=pd.bdate_range(Start_Date,T_Date)
```

```python
business_days
```

```python
print("Employee name:",Employee_Name,"Working days",len(business_days))
```
